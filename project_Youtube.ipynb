```python
video_id = 'rvd85gwzIRY'
def get_comments(video_id):
    comments = []
    request = youtube.commentThreads().list(
        part='snippet',
        videoId=video_id,
        maxResults=100
    )

    while request:
        response = request.execute()
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            comments.append(comment)

        if 'nextPageToken' in response:
            request = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                pageToken=response['nextPageToken'],
                maxResults=100
            )
        else:
            break
    return comments
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
for comment in comments:
    cleaned_comment = html.unescape(comment)
    sentiment_scores = analyzer.polarity_scores(cleaned_comment)


    if sentiment_scores['compound'] >= 0.05:
        sentiment = 'إيجابي'
    elif sentiment_scores['compound'] <= -0.05:
        sentiment = 'سلبي'
    else:
        sentiment = 'محايد'

    print(f"التعليق: {cleaned_comment} \nالتصنيف: {sentiment}")
```

```python
import matplotlib.pyplot as plt
```

```python
analyzer = SentimentIntensityAnalyzer()

positive_comments = 0
negative_comments = 0
neutral_comments = 0

for comment in comments:
    cleaned_comment = html.unescape(comment)
    sentiment_scores = analyzer.polarity_scores(cleaned_comment)


    if sentiment_scores['compound'] >= 0.05:
        positive_comments += 1
    elif sentiment_scores['compound'] <= -0.05:
        negative_comments += 1
    else:
        neutral_comments += 1

labels = ['positive', 'negative', 'neutral']
sizes = [positive_comments, negative_comments, neutral_comments]
colors = ['#4CAF50', '#FF5733', '#FFC300']

plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.axis('equal')
plt.title("Segment comments by sentiment")
plt.show()
```
