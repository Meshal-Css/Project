#**Python Implementation using the scikit-learn library**

#Step 1 : import libraries

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt                          # المكتبات المهمة
```

#Step 2 : Randomly chosen dataset

```python
 X = np.array([[5,3],
              [10,15],
              [15,12],
 ]) # تعرف متغير بسم
```

#Step 3 : Perform hierarchical clustering  + Ploting


```python
C = linkage(X, 'ward') # هذا السطر القصير هو المسؤول عن تنفيذ خوارزمية التجميع الهرمي على البيانات المُمثلة في المصفوفة X.
```

```python
dendrogram(C)
```

```python
dendrogram(C)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data point')
plt.ylabel('Distance')
plt.show()
```
